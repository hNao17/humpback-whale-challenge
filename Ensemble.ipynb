```python
import pandas as pd
import numpy as np
import json
```

```python
models = ['deskewed', 'cropped', 'v_deskewed', 'v_cropped']
leaderboard_scores = [0.38231, 0.38025, 0.41699, 0.41339]
```

```python
def get_results(path):
    with open(path) as f:
        json_data = f.read()
        res = json.loads(json_data)
    return res
```

```python
results = []

for m in models:
    res = get_results(f'./{m}_results.json')
    results.append(res)
```

```python
tot = sum(leaderboard_scores)
weights = [s / tot for s in leaderboard_scores]
```

```python
weights
```

```python
len(results[0])
```

```python
img_list = list(results[0].keys())
```

```python
preds_str = []
for i in range(len(results[0])):
    choices = []
    img = img_list[i]
    for m in range(len(results)):
        for choice in results[m][img]:
            choices.append((choice[0], (1 - choice[1]) * weights[m]))
    choices.sort(key=lambda x: x[1], reverse=True)
    best = []
    for c in choices:
        if c[0] not in best:
            best.append(c[0])
            if len(best) == 5:
                break
    preds_str.append(" ".join([x for x in best]))
```

```python
preds_str
```

```python
df = pd.DataFrame(preds_str, columns=["Id"])
df['Image'] = img_list
df.to_csv("ensemble_1.csv", index=False)
```
